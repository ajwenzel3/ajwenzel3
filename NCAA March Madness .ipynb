### Elo Rating explained

The Elo rating system is most commonly known for its use in chess, but as more sports introduce in-depth statistics and mathematical modeling into their respective leagues, new ways to rate and rank teams are being implemented. Elo works by using the difference between two team’s ratings as a prediction of the outcome. Each team’s rating will change accordingly to the actual outcome versus the predicted. For example, if a higher rated team beats a lower rated team, the expected outcome would have predicted such an outcome. Therefore, the higher rated team would have a small positive rating adjust and the lower rated team would have a small negative rating adjustment. This can work vice versa to benefit the “underdogs”. For our purposes in predicting the NCAA March Madness tournament, we found this rating system to be extremely useful in dealing with three areas.

<b>Win Percentage</b>: No matter if a team plays a worse rated team every single game, as long as they are winning games their rating will go up. However, Elo ratings expand on win percentage by giving more rating points to teams that beat tougher opponents. So a team from the Big Sky conference and the Big East conference might have the same win percentage but the Big East team should have a higher Elo rating as their competition throughout the season should be tougher. <br>
<br>
<b> Skill Gap</b>: Unlike other rating systems, the Elo system allows for us to easily quantify skill gaps between teams. Teams that consistently beat higher ranked opponents will be rewarded with a bigger elo rating adjustment. On the other hand, teams that beat lower ranked opponents will still recieve a positive adjustment, just not nearly as much as the underdog. 
<br>
<b>Recency Bias</b>: Elo takes into account if a team is either on a late season win streak or a slump. Teams on win streaks late in the season will have a higher Elo rating, which is perfect for helping predict tournament outcomes. If a lower <i> ranked </i> team on a win streak faces a blue chip school in a slump, Elo ratings may favor the lower ranked school. <br>
> For our Elo System we lowered the $K$ value that represents the value associated to each win. Recency Bias is important to have calculated in this model. However, some lower D1 schools with easier conferences and high win streaks had inflated ratings. Cutting the $K$ value in half allows us to keep recency bias a factor but limit the potential chance of ballooned ratings. 

<br>
Below is an equation of how the Elo rating system can predict the probability of a team winning.

$$\large E_a = \frac{1}{1 + 10^{\frac{(R_b-R_a)}{400}}} $$
<br>
This equation represents $E_a$ as the expected probability of team $A$ winning where $R_b$ and $R_a$ are the given ratings for both teams $A$ and $B$. The $400$ used in the denomonitor represents our scaler. It can be adjusted however we see want the expectation to correlate to the differences between each teams ratings.<br>
<br>
$$\large R_A' = R_A + K \cdot (S_A - E_A) $$
<br> This equation represents a team's new rating $R_A'$. It takes the old rating $R_A$ plus our $K$ value mentioned earlier times the actual score $S_A$ minus the expected score $E_A$.

### Changing Elo Based on KenPom Ratings

Based on prior tournaments data, solely predicting your bracket on Elo rating can be good but not great. We believe by just adding a few simple but effective conditions to our model will push us towards great. We will be adjusting using a multiplier corresponding to teams Adjusted Offensive and Defensive Effeciency KenPom Ratings. Normal efficiency ratings rate how many points are scored per 100 possessions: however, KenPom Adjusted ratings are based on how many possessions a team has per game. Allowing for slower paced teams to not be as negatively affected as they would be by normal effeciency ratings. 

KenPom is a ratings system originally created by Ken Pomeroy who is not affiliated with the NCAA, but is slowly becoming a household name as some stats using his rating methods are undeniably effective. As the tournament consists of more than just these seven teams, when considering how to adjust each teams elo rating dependent on their Adjusted Offensive and Defensive KenPom ratings, we decided that it would be best to take the median rank for each of the aforementioned ratings and proportionally reward or penalize team elo ratings based on their distnace from the median. Above average teams would get a positive elo rating multiplier adjustment while the inverse would be true for the below average teams. The equation used is listed below:

$$ \text{MarginRank}_a = ((\text{Median Offensive Rank}_{NCAA} - OffRank_a)+(\text{Median Defensive Rank}_{NCAA} - DefRank_a)) $$
$$ \text{AdjustedElo}_a = \text{BaseElo}_a \cdot (1 + (5*10^{-3})\cdot\text{Margin Rank}_a)$$

We chose a multiplier to adjust the Elo scores because after finding the rank margin, the only way for a team to get an extremely inflated Elo score would be for that team to be highly rated on both the offensive and defensive side of the ball. If a team is good on either end of the court, they deserve to have an Elo rating representitive to those numbers. We chose $5*10^{-3}$ because if a team happened to be extremely dominant on both ends and ranked $1^{st}$ in each rating, their score would only be increased by ~$88$%. From past data and looking at ratings for this years teams, that number is almost impossible to reach.

```python
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#get teams

url = "https://www.ncaa.com/standings/basketball-men/d1"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

if (page.status_code == 200):
    ts = soup.find_all("td", class_="standings-team")

teams = [None]*len(ts)  #empty array of teams

print(f"Found {len(teams)} teams.")

for i in range(0,len(teams)):
    teams[i] = ts[i].text
```

```python
#loop over season dates and extract games


years = [2022,2022,2022,2023,2023, 2023]
months = ["10", "11", "12", "01", "02", "03"]
days = np.arange(1,32)

gamectr = 0
games = []

for j in range(0,len(months)):
    for day in days:
        if day < 10:
            url = f"https://www.ncaa.com/scoreboard/basketball-men/d1/{years[j]}/{months[j]}/0{day}"
        else:
            url = f"https://www.ncaa.com/scoreboard/basketball-men/d1/{years[j]}/{months[j]}/{day}"
        page = requests.get(url)
        if (page.status_code == 200):
            soup = BeautifulSoup(page.content, "html.parser")
                       
            names = soup.find_all("span", class_="gamePod-game-team-name")
            scores = soup.find_all("span", class_="gamePod-game-team-score")
            
            N = int(len(names)/2) #number of games
            
            for i in range(0,int(len(names)/2)):
                if ((names[2*i].text in teams) and (names[2*i+1].text in teams)):  #make sure both teams are D1
                    #print(f"{names[2*i].text} is in teams and so is {names[2*i+1].text}")
                    #gamectr += 1
                    if not((scores[2*i].text == "") or (scores[2*i+1].text == "")): #make sure both teams have a score, ie game played
                        gamectr += 1
                        games.append([names[2*i].text,names[2*i+1].text,scores[2*i].text,scores[2*i+1].text,years[j],months[j],day])
```

```python
#Temporary array to store KenPom Ratings
ratingsDataArray = []
#Get csv file
with open("./MarchMadnessRatings_Standardized.csv", 'rt') as f:
    for line in f.readlines()[1:]:
        parts =line.strip().split(",")
        if parts[0] != "2023":
            continue
        name = parts[1]
        offRating = parts[3]
        offRank = parts[4]
        defRating = parts[5]
        defRank = parts[6]
        ratingsDataArray.append(parts)
#Reordered array to share the same indeces as the elo array
rawRatingsData = np.array(ratingsDataArray)
ratingsData = np.empty_like(rawRatingsData)
for ratings in rawRatingsData:
    team_name = ratings[1]
    index = teams.index(team_name)
    ratingsData[index] = ratings
```

```python
#Base Elo system

numgames= len(games) 
#For every 400 points difference between the ratings it is 10x as likely for the higher rated team to win and vice versa
base_scale = 400
#K represents the adjustment per game. We reduced this number to combat inflated recency bias.
K = 15
names = teams
elo = np.full(len(teams),1000)

for i in range(numgames-1,0,-1):
    hteam = games[i][1]
    ateam = games[i][0]
    hscore = games[i][3]
    ascore = games[i][2]
    
    #get ratings
    hrat = elo[names.index(hteam)]
    arat = elo[names.index(ateam)]
    diff = hrat - arat
    
    
    #calc expected results/probs of winning
    hexp = 1./(1. + 10**(diff/base_scale))
    aexp = 1. - hexp
    
    #update ratings
    if (hscore > ascore): #home team win
        elo[names.index(hteam)] += K*(1. - hexp)
        elo[names.index(ateam)] += K*(0. - aexp)
    elif (ascore > hscore): #away team win
        elo[names.index(hteam)] += K*(0. - hexp)
        elo[names.index(ateam)] += K*(1. - aexp)
    #no else or draws needed 

#Sorted Base Elo Rating Highest to Lowest
mappedElo = { names[i] : elo[i] for i in range(len(names))}
[print(f"{pair[0]} - {pair[1]}") for pair in sorted(mappedElo.items(), key=lambda x : x[1], reverse=True)]
```

```python
#Used for taking the difference of teams rank versus median rank for both offensive and defensive ratings.
margin = 0
#Grabs raw ratings into a convenient array
offRanks = np.array(ratingsData[:, 4])
defRanks = np.array(ratingsData[:, 6])
#Our new adjusted elo rating
adjustedElo = np.zeros_like(elo)


for name in names:
    i = names.index(name)
    #Takes in median rank of teams offensive and defensive ratings
    #Subtract specific teams offensive and defensive ranks from the median to get the teams margin
    margin = ((len(teams)/2)-float(offRanks[i])) + ((len(teams)/2)-float(defRanks[i]))
    #Adjusted Elo takes into account the original elo times a multiplier based on the margin
    adjustedElo[i] = elo[i]*(1 + margin*0.0005)

#Sorted Adjusted Elo Rating Highest to Lowest
mappedElo = { names[i] : adjustedElo[i] for i in range(len(names))}
[print(f"{pair[0]} - {pair[1]}") for pair in sorted(mappedElo.items(), key=lambda x : x[1], reverse=True)]
```
